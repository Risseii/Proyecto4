```
import requests
from PIL import Image
from urllib.request import urlopen
import json
import os

pokemon = input("Ingresar el nombre de un pokemon")

url = 'https://pokeapi.co/api/v2/pokemon/'+ pokemon
respuesta = requests.get(url)

#En caso no se encuentre el pokemon, se genere una respuesta que no se encontró
if respuesta.status_code != 200:
  print("Pokemon no encontrado")
  exit()

#A partir de una variables datos acceder al peso, tamaño, movimiento,habilidades, tipos y la img
datos = respuesta.json()

print("Peso:")
print(datos['weight'])

print("Tamaño:")
print(datos['height'])

print("Movimientos:")
for i in range(len(datos['moves'])):
    print(datos["moves"][i]['move']['name'])

print("Habilidades:")
for i in range(len(datos['abilities'])):
    print(datos["abilities"][i]['ability']['name'])

print("Tipos:")
for i in range(len(datos['types'])):
    print(datos['types'][i]['type']['name'])

print("Imagen:")
url_imagen = datos['sprites']['other']['home']['front_default']
imagen = Image.open(urlopen(url_imagen))
imagen.show()

#Se crea un diccionario vacio info y se agrega la información
info = {}
info['Peso'] = datos['weight']
info['Tamaño'] = datos['height']
info['Movimientos'] = [move['move']['name'] for move in datos['moves']]
info['Habilidades'] = [ab['ability']['name'] for ab in datos['abilities']]
info['url_imagen'] = datos['sprites']['other']['home']['front_default']
info['tipo'] = [tp['type']['name'] for tp in datos['types']]
#print(info) 


#Se guarda el diccionario info en un archivo.json
with open('pokedex.json','w') as archivo:
    json.dump(info,archivo)
```
